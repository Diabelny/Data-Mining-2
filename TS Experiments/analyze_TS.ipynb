```python
# Read the first TS of Ravdess Audio File Folder with emotional intensity=normal
path='../Ravdess Audio file/03-01-01-01-01-01-01.wav' 
ts, sr = librosa.load(path, sr=None)
print(ts.shape, sr) # time series with 158558 points? and sampling rate 48000?
```

```python
plt.plot(ts) # x=time y=value
plt.show()
```

```python
Audio(ts, rate=sr)
```

```python
# downsample the audio
q = 8
ts1 = decimate(ts[50000:58000], q=8)
Audio(ts1, rate=sr/q)
```

```python
plt.plot(ts1) # only the word kids
plt.show()
```
